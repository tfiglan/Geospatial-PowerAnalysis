## Calculate Load Demand

### Subtask:
Calculate the active power (kW), reactive power (kVAr), and apparent power (kVA) for each individual load across 'Peak', 'Medium', and 'Minimum' loading scenarios.


**Reasoning**:
The subtask requires calculating the active power (kW), reactive power (kVAr), and apparent power (kVA) for each individual load across different loading scenarios. This code block will perform these calculations and store the results in a list of dictionaries.



````python
import math

# 1. Initialize an empty list to store all load data for all scenarios
all_scenario_load_data = []

power_factor = 0.95 # Given power factor

# 2. Iterate through each loading scenario
for scenario_name, data in loading_scenarios_e2.items():
    loading_factor = data['factor']

    # 3. For each scenario, iterate through every load
    for load_name, peak_kw in original_peak_load_kW_e2.items():
        # 4. Calculate current_kw for each load
        current_kw = peak_kw * loading_factor

        # 5. Calculate current_kvar
        # kVAr = kW * tan(arccos(PowerFactor))
        # Handle the case where power_factor is 1 to avoid math domain error in acos
        if power_factor == 1:
            current_kvar = 0.0
        else:
            # Ensure the argument for acos is within [-1, 1]
            pf_clamped = max(-1.0, min(1.0, power_factor))
            current_kvar = current_kw * math.tan(math.acos(pf_clamped))

        # 6. Calculate current_kVA
        current_kVA = math.sqrt(current_kw**2 + current_kvar**2)

        # 7. Store the results
        all_scenario_load_data.append({
            'Scenario': scenario_name,
            'Load Name': load_name,
            'kW': round(current_kw, 2),
            'kVAr': round(current_kvar, 2),
            'kVA': round(current_kVA, 2)
        })

# Convert the list of dictionaries to a pandas DataFrame
df_all_scenario_loads = pd.DataFrame(all_scenario_load_data)

print("Calculated load demand for each load across all scenarios:")
print(df_all_scenario_loads.head())
print(df_all_scenario_loads.tail())
````

**Reasoning**:
The subtask requires transforming the voltage data from wide format (one column per bus) to long format (one row per bus-phase-scenario combination). This will make the data suitable for visualization and comparison. I will create a helper function to process each DataFrame and handle both single-phase and three-phase voltage representations, then combine the results.



````python
processed_voltage_data = []

# Get all bus columns by excluding 'Scenario' and 'Compliant'
bus_columns = [col for col in df_voltages_e2.columns if col not in ['Scenario', 'Compliant']]

def process_dataframe(df, pv_status_label):
    for index, row in df.iterrows():
        scenario = row['Scenario']
        for col_name in bus_columns:
            bus_base = col_name.replace(' (pu)', '') # Extract bus name without '(pu)'
            voltage_value = row[col_name]

            if isinstance(voltage_value, str) and 'P1:' in voltage_value: # Three-phase bus
                phases = voltage_value.split(', ')
                for phase_entry in phases:
                    parts = phase_entry.split(':')
                    phase_name = parts[0]
                    voltage_pu = float(parts[1])
                    processed_voltage_data.append({
                        'Scenario': scenario,
                        'Bus': bus_base,
                        'Phase': phase_name,
                        'Voltage (pu)': voltage_pu,
                        'PV Status': pv_status_label
                    })
            else: # Single-phase bus
                voltage_pu = float(voltage_value)
                processed_voltage_data.append({
                    'Scenario': scenario,
                    'Bus': bus_base,
                    'Phase': 'All',
                    'Voltage (pu)': voltage_pu,
                    'PV Status': pv_status_label
                })

# Process df_voltages_e2 (No PV)
process_dataframe(df_voltages_e2, 'No PV')

# Process df_voltages_e2_pv (With PV)
process_dataframe(df_voltages_e2_pv, 'With PV')

# Convert the list to a DataFrame
df_all_voltages_long = pd.DataFrame(processed_voltage_data)

print("Transformed voltage data (first 5 rows):")
print(df_all_voltages_long.head().to_markdown(index=False))
print("Transformed voltage data (last 5 rows):")
print(df_all_voltages_long.tail().to_markdown(index=False))
````
